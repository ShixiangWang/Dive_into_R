## 第一步

让我们使用第一个图来回答这样一个问题：有大引擎的汽车是不是比小引擎汽车消耗燃料多？你可能心中已经有了答案，但尝试更加精确地回答这个问题。引擎的大小和燃料效率是怎样的关系？是正相关？负相关？线性相关？非线性相关？

### `mpg`数据框

你可以使用`mpg`数据框来检验你的答案。**数据框是一个以列为变量、行为观测值的表格形式数据集合**。`mpg`包含了US环境保护局收集的38种模型车观测数据。

````r
ggplot2::mpg # 注意jupyter notebook显示与rstudio有些不同
````

- `displ`是汽车的引擎大小，以升为单位
- `hwy`是汽车在高速公路上的燃油效率，以英里每加仑为单位。

想获取更多关于`mpg`信息，运行`?mpg`打开帮助页面。

### 创建一个ggplot对象

想要绘制`mpg`，运行下面的代码令`displ`在x轴上，`hwy`在y轴上：

````r
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy))
````

图形显示引擎大小和燃油效率呈现负相关。

**`ggplot()`创建了一个可以添加图层的坐标系统。**函数的第一个参数是数据集，因此使用`ggplot(data = mpg)`可以创建一个空图，我们可以通过添加一个或多个图层到`ggplot()`来完成绘图。函数`geom_point()`添加一个点图的图层，它会创建一个散点图。ggplot2有非常多的几何函数用来添加不同类型的图层。

ggplot2里面每一个几何函数都有一个`mapping`参数，它定义了数据集里面的变量怎样映射到可视化的属性。`mapping`参数经常放在`aes()`函数里，`aes()`函数里的`x`和`y`指定了那个变量映射到x和y轴。ggplot2会在`data`参数里寻找映射的变量，这里是`mpg`。

**简单地理解，ggplot()初始化了画图的画板，并指出使用什么作为数据集，然后通过不同的几何函数在画板上绘制不同的图形，图形的数据来自数据集的变量。**

### 一个绘图模版

让我们把代码变成可以重复利用的模板：

```R
ggplot(data = <DATA>) +
    <GEOM_FUNCTION>(mapping = aes(MAPPINGS>))
```

使用时在相应位置填入数据集，变量和函数。

## 美学映射

> 一张图片的价值是迫使我们注意到那些意料之外的事情 —— John Tukey

建议读一读[ggplot2入门与进阶（上）](http://www.cellyse.com/how_to_use_gggplot2_part1/)以及下部分。

我们可以在二维数据上添加一些其他的图形属性，比如形状、大小、颜色等等。

比如我们下面将不同的汽车类别以不同颜色表示：

````r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy, color = class))
````

````r
# 映射到大小
ggplot(data = mpg) + geom_point(mapping = aes(x=displ, y=hwy, size=class))
````

````r
# 或者我们可以使用透明度 
ggplot(data = mpg) + geom_point(mapping = aes(x=displ, y=hwy, alpha=class))
````

````r
# 或者使用形状
ggplot(data = mpg) + geom_point(mapping = aes(x=displ, y=hwy, shape=class))
````

我们还可以手动设定标度（比如颜色）


````r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy), color="blue")
````

我们需要理解标度的一些设定：

1. 颜色的名字就是字符串（英文）
2. 点的大小以mm为单位
3. 点的形状以数字显示，如下

![](http://r4ds.had.co.nz/visualize_files/figure-html/shapes-1.png)

## 分面

另一种用分类变量添加映射的有用方式是分面，把一个图安装不同分类分割为不同的子图。

如果根据一个变量分面，使用`facet_wrap()`函数，该函数的参数是公式，比如我们按`class`把mpg数据集画出来

````r
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy)) +
    facet_wrap( ~ class, nrow=2)
    
````

如果有两个分类变量组合，使用`facet_grid()`函数

````r
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy)) +
    facet_grid(drv ~ cyl)
````

````r
# 使用`.`可以替换变量名
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy)) +
    facet_grid(. ~ cyl)
````

## 几何对象

下面两幅图相似吗？

![](http://r4ds.had.co.nz/visualize_files/figure-html/unnamed-chunk-18-1.png) ![](http://r4ds.had.co.nz/visualize_files/figure-html/unnamed-chunk-18-2.png)

每个图都使用了不同的可视化对象去呈现数据，用ggplot2语法来说，它们使用了不同的**几何对象**。

想要改变图形中的几何对象，我们要更改添加到`ggplot()`函数的几何函数。比如，想要绘制上面两幅图，使用以下代码：

````r
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy))
````

````r
ggplot(data = mpg) + 
    geom_smooth(mapping = aes(x = displ, y = hwy))
````

每个几何函数都有`mapping`参数，但并不是所有标度（x,y，大小，颜色等等）都可以用。比如，你可以给点设置形状，但你不能给线设置形状。

从另一个角度讲，用不同的线条形状，你可以绘制不同的线条。

````r
ggplot(data = mpg) + 
    geom_smooth(mapping = aes(x = displ, y = hwy, linetype = drv))
````

让我们看看更多的图形设置效果

````r
ggplot(data = mpg) +
  geom_smooth(mapping = aes(x = displ, y = hwy))
              
ggplot(data = mpg) +
  geom_smooth(mapping = aes(x = displ, y = hwy, group = drv))
    
ggplot(data = mpg) +
  geom_smooth(
    mapping = aes(x = displ, y = hwy, color = drv),
    show.legend = FALSE
  )
````

如果想在同一张图显示多个几何对象，可以叠加多个几何函数（**注意添加顺序和最后的显示效果相关，用多了你就明白了，嘿嘿嘿**）

````r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) +
  geom_smooth(mapping = aes(x = displ, y = hwy))
````

你可以直接在`ggplot()`函数里面设置映射关系，后续几何函数直接添加其他参数就可以了（甚至不要任何参数）

````r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
  geom_point() + 
  geom_smooth() # ggplot()里面设置的相当于全局参数
````

而你在几何函数里设置的参数被认为是局部参数，它可以重写全局参数。

````r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
  geom_point(mapping = aes(color = class)) + 
  geom_smooth()
````

你可能已经隐约意识到，不同的图层可以使用不同的数据，没错

````r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
  geom_point(mapping = aes(color = class)) + 
  geom_smooth(data = filter(mpg, class == "subcompact"), se = FALSE)
````

## 统计变换


Many graphs, like scatterplots, plot the raw values of your dataset. Other graphs, like bar charts, calculate new values to plot:

    bar charts, histograms, and frequency polygons bin your data and then plot bin counts, the number of points that fall in each bin.

    smoothers fit a model to your data and then plot predictions from the model.

    boxplots compute a robust summary of the distribution and then display a specially formatted box.


在已有基础之上，为图形计算新的数值叫做**统计变换**，下面图形描述了`geom_bar()`怎么处理这一过程

![](http://r4ds.had.co.nz/images/visualization-stat-bar.png)

函数中的`stat`参数指定了使用什么样的统计函数进行处理，像`geom_bar`的默认`stat`值为`count`，意思是使用`stat_count()`函数。

我们可以将两种函数交换使用，比如用`stat_count()`替换`geom_bar()`

````r
ggplot(data = diamonds) + 
  stat_count(mapping = aes(x = cut)) # 和使用geom_bar()结果一致
````

````r
# 我们可以修改stat参数来绘制不同性质的直方图

demo <- tribble(
  ~cut,         ~freq,
  "Fair",       1610,
  "Good",       4906,
  "Very Good",  12082,
  "Premium",    13791,
  "Ideal",      21551
)

ggplot(data = demo) +
  geom_bar(mapping = aes(x = cut, y = freq), stat = "identity")
````

````r
# 你可能想用比率替换计数

ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, y = ..prop.., group = 1))
````

````r
# 你可能想展示一些特征性统计值
ggplot(data = diamonds) + 
  stat_summary(
    mapping = aes(x = cut, y = depth),
    fun.ymin = min,
    fun.ymax = max,
    fun.y = median
  )
````

ggplot2提供了超过20种统计函数，强大无比。

## 位置调整

直方图更有用的标度是`fill`而不是直接使用`color`

````r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, colour = cut))
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, fill = cut))
````

如果你把`fill`赋给一个不同的变量会发生什么？

````r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, fill = clarity))
````

直方图会自动进行堆叠。

堆叠是通过`position`参数指定的位置调整之一。如果你不想要堆叠效果，可以使用其他三个参数之一："identity"，"dodge" 或 "fill"

- `position="identity"会准确的按你的设置放置几何对象，在直方图中通常不是很有用，因为会产生覆盖。下面是例子

````r
ggplot(data = diamonds, mapping = aes(x = cut, fill = clarity)) + 
  geom_bar(alpha = 1/5, position = "identity")
ggplot(data = diamonds, mapping = aes(x = cut, colour = clarity)) + 
  geom_bar(fill = NA, position = "identity")
````

这个参数在二维几何对象非常有用，像点图就是默认配置。

- `position="fill"`看起来像堆叠，但会使每个堆叠图有相同的高度，这让比较组间的比例非常方便。

````r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, fill = clarity), position = "fill")
````

- `position="dodge"`把重叠的对象都一个一个挨着放置，这让比较组内数据更加方便。

````r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, fill = clarity), position = "dodge")
````

其实还有一个参数，在绘制散点图时非常有用。当图中点很多时，点会堆积重叠，图形就不好看了，我们可以设置`position="jitter"`来对点添加少量的抖动。

![](http://r4ds.had.co.nz/visualize_files/figure-html/unnamed-chunk-42-1.png)

````r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy), position = "jitter")
````

更多调整信息，R中键入

```
?position_dodge, ?position_fill, ?position_identity, ?position_jitter, and ?position_stack.
```

## 坐标系统

坐标系统可能是ggplot2最复杂的一部分。默认的坐标系统是笛卡尔坐标系，x与y的位置决定了点的位置。也由一些其他的坐标系，有时候会有用。

- `coord_flip()`反转x,y轴。这个有时候很有用

````r
ggplot(data = mpg, mapping = aes(x = class, y = hwy)) + 
  geom_boxplot()
ggplot(data = mpg, mapping = aes(x = class, y = hwy)) + 
  geom_boxplot() +
  coord_flip()
````

- `coord_quickmap()`为地图设置正确的比例。如果你用ggplot2绘制空间数据这是非常重要的。

````r
nz <- map_data("nz")

ggplot(nz, aes(long, lat, group = group)) +
  geom_polygon(fill = "white", colour = "black")

ggplot(nz, aes(long, lat, group = group)) +
  geom_polygon(fill = "white", colour = "black") +
  coord_quickmap()
````

- `coord_polar()`使用极坐标系

````r
bar <- ggplot(data = diamonds) + 
  geom_bar(
    mapping = aes(x = cut, fill = cut), 
    show.legend = FALSE,
    width = 1
  ) + 
  theme(aspect.ratio = 1) +
  labs(x = NULL, y = NULL)

bar + coord_flip()
bar + coord_polar()
````

## 图形的图层语法

在前面一些部分，我们学习了如何绘制散点、直方图以及箱线图等等。之前我们也提及了绘图的模板，经过这些部分的学习，我们是时候更新它了。

```
ggplot(data = <DATA>) + 
  <GEOM_FUNCTION>(
     mapping = aes(<MAPPINGS>),
     stat = <STAT>, 
     position = <POSITION>
  ) +
  <COORDINATE_FUNCTION> +
  <FACET_FUNCTION>
```

我们的新模板有7个参数（尖括号括起来的单词）。在实际操作中，ggplot2已经提供了非常有用的默认选项，所以我们很少会需要补充这所有的7个参数。

**模板里面的7个参数组成了图形语法，它是构建图形的标准系统**。图形语法基于你可以通过**数据集**，**几何对象**，**映射集**，**统计变换**，**位置调整**，**坐标系统**以及**分面主题**的组合来描绘任意一个图形。
